# Data Preprocessing

```python
data1=data
data1.head()
```

The data from column 'cabin' is mostly missing. So this column is dropped.

```python
data1.drop('Cabin',axis=1,inplace=True)
```

```python
data1["Age"] = data1["Age"].fillna(data1["Age"].mean())
data1['Embarked']= data1['Embarked'].fillna(data1['Embarked'].mode()[0])
```

```python
data1.head()
```

```python
data.isna().sum()
```

Handling outliers

```python
data1.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
```

```python
data1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data1['Sex']= le.fit_transform(data1['Sex'])
data1['Embarked']= le.fit_transform(data1['Embarked'])
```

```python
data1.head()
```

```python
x=data1.drop('Survived',axis=1)
y=data1['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score, precision_score,recall_score, f1_score
```

Naive Bayes Algorithm

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train,y_train)
y_predgnb = gnb.predict(x_test)
y_train_predgnb= gnb.predict(x_train)
print('Accuracy Score is:', accuracy_score(y_test, y_predgnb))
print('Recall Score is:', recall_score(y_test, y_predgnb))
print('Precision Score:', precision_score(y_test, y_predgnb))
print('F1 score is:', f1_score(y_test, y_predgnb))
print("\n The confusion matrix is : \n", confusion_matrix(y_test, y_predgnb))
```

Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc= DecisionTreeClassifier()
dtc.fit(x_train,y_train)
y_preddtc= dtc.predict(x_test)
y_train_preddtc = dtc.predict(x_train)
print("Training Accuracy is : ", accuracy_score(y_train, y_train_preddtc))
print('Accuracy Score is:', accuracy_score(y_test, y_preddtc))
print('Recall Score is:', recall_score(y_test, y_preddtc))
print('Precision Score:', precision_score(y_test, y_preddtc))
print('F1 score is:', f1_score(y_test, y_preddtc))
print("\n The confusion matrix is : \n", confusion_matrix(y_test, y_preddtc))
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
y_predrfc= rfc.predict(x_test)
y_train_predrfc= rfc.predict(x_train)

print("Training Accuracy is : ", accuracy_score(y_train, y_train_predrfc))
print('Accuracy Score is:', accuracy_score(y_test, y_predrfc))
print('Recall Score is:', recall_score(y_test, y_predrfc))
print('Precision Score:', precision_score(y_test, y_predrfc))
print('F1 score is:', f1_score(y_test, y_predrfc))
print("\n The confusion matrix is : \n", confusion_matrix(y_test, y_predrfc))
```

Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_predlr = lr.predict(x_test)
y_train_predlr = lr.predict(x_train)
print("Training Accuracy is : ", accuracy_score(y_train, y_train_predlr))
print('Accuracy Score is:', accuracy_score(y_test, y_predlr))
print('Recall Score is:', recall_score(y_test, y_predlr))
print('Precision Score:', precision_score(y_test, y_predlr))
print('F1 score is:', f1_score(y_test, y_predlr))
print("\n The confusion matrix is : \n", confusion_matrix(y_test, y_predlr))
```

KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
knnc = KNeighborsClassifier(n_neighbors=4,metric = "minkowski") 
knnc.fit(x_train,y_train)
y_predknnc = knnc.predict(x_test)
y_train_predknnc = knnc.predict(x_train)
print("Training Accuracy is : ", accuracy_score(y_train, y_train_predknnc))
print('Accuracy Score is:', accuracy_score(y_test, y_predknnc))
print('Recall Score is:', recall_score(y_test, y_predknnc))
print('Precision Score:', precision_score(y_test, y_predknnc))
print('F1 score is:', f1_score(y_test, y_predknnc))
print("\n The confusion matrix is : \n", confusion_matrix(y_test, y_predknnc))
```
